```python
    def decSesionesRepetidas(func):
        def wrap(self):
            repetidas= self.repeatedSes()
            sessionF=self.filtSes(repetidas)
            return func(sessionF)
        return wrap
```

```python
from myDf import *
a=myData()
a.loadDf_fromExcel()
a.getStyledInf()
```

```python
repetidas= a.repeatedSes()
sessionF=a.filtSes(repetidas)
a.infSesFil()
```

```python
sesSiones=[]
repetidas=[]

for indx in a.session.index:
    sesName = a.session.loc[indx]["Nombre"]
    if sesName in sesSiones:
        repetidas.append(sesName)
    sesSiones.append(sesName)



for i,indx  in enumerate(a.session.index):

    sesName = a.session.loc[indx]["Nombre"]

    if sesName  in repetidas:
        sesSiones[i]=sesName+f" ({a.session.loc[indx]["Creado"].strftime("%Y-%m-%d %H:%M:%S")})"
        #print(sesSiones[i])

sessionF=a.session.copy()
sessionF["Nombre"] = sesSiones
sessionF
```

```python
df=a.df.copy()
ses=[sessionF.loc[indx]["Nombre"] for indx in df["Sesión"].to_list()]
df["Sesión"]=ses
df
```

```python

md=[]
for i in a.df["Sesión"]:
    sesion=a.session.loc[i]
    md.append(sesion["matchDay"])
```

```python
d=a.df_st.copy()
d["MD"]=md
pd.unique(d["MD"])
```
